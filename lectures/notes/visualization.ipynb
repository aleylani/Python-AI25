## Matplotlib

Matplotlib är en av de vanligaste paketen som används i Python för att skapa diagram och visualiseringar av data.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate some data

x = np.linspace(0, 10, 50)
y_1 = 4 + np.sin(2*x)
y_2 = 4 + np.cos(4*x)

plt.plot(x, y_1) # plotta direkt via plt.plot, detta passar bra om vi endast ska plotta en figur
plt.plot(x, y_2, color='tab:pink', marker='o', alpha=0.3)

plt.title('Mina exempelplottar', fontsize=20) # lägg till en titel, fontsize kontrollerar teckenstorleken
plt.xlabel('x')
plt.ylabel('y', rotation=0, labelpad=15)

plt.show()
```

En scatterplot är som en lineplot, men där vi inte drar en linje mellan varje på varandra följande punkt - utan endsat markerar givna (x,y) koordinater

```python
x = np.linspace(0, 10, 50)
y_1 = 4 + np.sin(2*x)
y_2 = 4 + np.cos(4*x)

plt.scatter(x, y_1)
plt.scatter(x, y_2, color='pink', marker='*', alpha=0.6)

plt.title('Mina exempelplottar', fontsize=20) # lägg till en titel, fontsize kontrollerar teckenstorleken
plt.xlabel('x')
plt.ylabel('y', rotation=0, labelpad=15)

plt.show()
```

Vi använder *figure* för att kontrollera hela "tavlan" i stort.

```python
x = np.linspace(0, 10, 50)
y_1 = 4 + np.sin(2*x)
y_2 = 4 + np.cos(4*x)

plt.figure(figsize=(8,5))  # ange storlek 8x5 tum, figsize=(8,5) anger att figuren ska vara 8 tum bred och 5 tum hög

plt.scatter(x, y_1, label='f(x)')
plt.scatter(x, y_2, color='pink', marker='*', alpha=0.6, label='g(x)')

plt.title('Mina exempelplottar', fontsize=10) # lägg till en titel, fontsize kontrollerar teckenstorleken
plt.xlabel('x')
plt.ylabel('y', rotation=0, labelpad=15)

plt.xticks(rotation=45)   # rotera x-axelns etiketter med 45 grader
plt.xticks(ticks=[0,2,4,6,8,10], labels=['noll', 'två', 'fyra', 'sex', 'åtta', 'tio'])

plt.legend(loc='upper right')   # lägg till en förklarande låda för labels, loc='upper right' anger att den ska vara i det övre högre hörnet
 
plt.show()  # jätteviktigt att alltid avsluta våra plottar med plt.show(), annars kommer de inte synas om vi lägger in de
            # i en funktion eller i en loop, eller kör via ett script.
```

## Subplots

```python
x = np.linspace(0, 10, 50)
y_1 = 4 + np.sin(2*x)
y_2 = 4 + np.cos(4*x)
y_3 = [i + 3 for i in x]

# med axes kan du kontrollera varje enskild graf i vår plot
# med fig kontrollerar du övergripane inställningar för alla dina plottar

fig, axes = plt.subplots(1, 3, figsize=(14,4))

fig.suptitle('Mina stora titel', fontsize=20)

axes[0].plot(x, y_1, label='f(x)')
axes[1].scatter(x, y_2, label='g(x)', color='pink', marker='*', alpha=0.6) 
axes[2].plot(x, y_3, label='h(x)', color='green')

axes[0].set_title('Subplot 1')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axes', rotation=0, labelpad = 20)
axes[0].legend(loc='upper right')  # lägg till en förklarande låda på varje subplot, loc='upper right' anger att den ska vara i det övre högra hörnet

plt.tight_layout()

plt.show()

```

```python
# Generate dummy data
x = np.linspace(0, 10, 50)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x) / 5
y_4 = np.exp(-x)
y_5 = np.log(x + 1)
y_6 = x ** 2

# Create a 2x3 subplot grid

fig, axes = plt.subplots(2, 3, figsize=(15, 8)) # 2 rows, 3 columns


# Plot each dataset in its respective subplot
axes[0, 0].plot(x, y_1, color='b', label='sin(x)')
axes[0, 0].scatter(x, y_1+1, color='r', label='sin(x)+1')
axes[0, 0].set_title('Subplot 1')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')
axes[0, 0].legend()
axes[0, 0].grid(True)

axes[0, 1].plot(x, y_2, color='g', label='cos(x)')
axes[0, 1].set_title('Subplot 2')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y')
axes[0, 1].legend()
axes[0, 1].grid(True)

axes[0, 2].plot(x, y_3, color='r', label='tan(x)/5')
axes[0, 2].set_title('Subplot 3')
axes[0, 2].set_xlabel('x')
axes[0, 2].set_ylabel('y')
axes[0, 2].legend()
axes[0, 2].grid(False)

axes[1, 0].plot(x, y_4, color='c', label='exp(-x)')
axes[1, 0].set_title('Subplot 4')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
axes[1, 0].legend()
axes[1, 0].grid(True)

axes[1, 1].plot(x, y_5, color='m', label='log(x+1)')
axes[1, 1].set_title('Subplot 5')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('y')
axes[1, 1].legend()
axes[1, 1].grid(True)

axes[1, 2].plot(x, y_6, color='y', label='x^2')
axes[1, 2].set_title('Subplot 6')
axes[1, 2].set_xlabel('x')
axes[1, 2].set_ylabel('y')
axes[1, 2].legend()
axes[1, 2].grid(True)

plt.tight_layout()

plt.show()
```
