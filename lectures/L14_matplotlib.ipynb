```python
def f(x):

    y = x**2

    return y

def g(x):

    y = x + 2

    return y
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = [x for x in range(-10, 11)]
ypoints = [f(x) for x in range(-10, 11)]

plt.plot(xpoints, ypoints, marker='*', color='green', linestyle = 'dashed')

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.ylabel('f(x)', fontdict = font2, rotation = 0)
plt.xlabel('x', fontdict = font2)
plt.title('The graph of y = x^2', fontdict = font1)

plt.show()
```

```python
xpoints = [x for x in range(-10, 11)]
fpoints = [f(x) for x in range(-10, 11)]
gpoints = [g(x) for x in range(-10, 11)]

plt.plot(xpoints, fpoints, marker='*', color='green', linestyle = 'dashed')
plt.title('f(x)')
plt.show()

plt.plot(xpoints, gpoints)
plt.title('g(x)')
plt.show()
```
